```python
import pandas as pd
import matplotlib.pyplot as plt
```

Crear un DataFrame


```python
years = [2012, 2013, 2014, 2015]
califications = [100, 96, 75, 30]

calif_dict = {'years': years, 'califications': califications}

print(calif_dict)
```

```python
cal_df = pd.DataFrame(calif_dict)

cal_df.head()
```

```python
plt.plot('years', 'califications', data=cal_df)
plt.title('Calificaciones a lo largo de los años')
plt.xlabel('Años')
plt.ylabel('Calificaciones')
plt.show()
```

```python
netflix_df_movies = netflix_df[netflix_df['type']== 'Movie']
netflix_df_movies['type']. unique()
```

```python
netflix_df_movies.head()
```

```python
netflix_df_movies_columns = netflix_df_movies[['title', 'country', 'rating', 'release_year', 'duration']]
netflix_df_movies_columns.head()
```

Gráfico de Puntos

```python
fig = plt.figure(figsize=(12,8))
plt.scatter('release_year', 'duration', data=netflix_df_movies_columns)
plt.xlabel('Años')
plt.ylabel('Minutos')
plt.title('Duración de las peliculas')
plt.show()
```

```python
netflix_df_movies_columns.nsmallest(n=1, columns='duration')
```

```python
netflix_df_movies_columns.nlargest(n=1, columns='duration')
```

```python
netflix_df_movies_columns['rating'].value_counts()
```

```python
netflix_df_movies_filtered = netflix_df_movies_columns[
    (netflix_df_movies_columns['rating'] == 'TV-MA') |
    (netflix_df_movies_columns['rating'] == 'TV-14') |
    (netflix_df_movies_columns['rating'] == 'TV-PG')
]

netflix_df_movies_filtered['rating'].unique()
```

```python
colors = []

for lab, row in netflix_df_movies_filtered.iterrows():
  if row['rating'] == 'TV-MA':
    colors.append('red')
  elif row['rating'] == 'TV-14':
    colors.append('blue')
  else:
    colors.append('black')

  print(colors[:10])
```

```python
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12,8))

plt.scatter('release_year', 'duration', data=netflix_df_movies_filtered, color=colors)
plt.xlabel('Años')
plt.ylabel('Minutos')
plt.title('Duración de las peliculas TV-MA, TV-14 y TV-PG')
plt.show()
```

```python
count_movies = netflix_df_movies_columns['rating'].value_counts()
```

```python
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12,8))

plt.bar(count_movies.index, count_movies.values)
plt.title('Cantidad de peliculaspor género')
plt.xlabel('Género')
plt.ylabel('Cantidad')
plt.xticks(rotation=90)
plt.show()
```

```python
movies_duration_year = netflix_df_movies_columns.groupby('release_year')['duration'].mean().reset_index()
```

```python
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12,8))

plt.plot('release_year', 'duration', data=movies_duration_year)
plt.title('Duración de las peliculas por año')
plt.xlabel('Años')
plt.ylabel('Duración promedio')
plt.show()
```

```python
 count_countries = netflix_df_movies_columns['country'].value_counts().head(10)
```

```python
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12,8))

plt.barh(count_countries.index, count_countries.values)
plt.title('Paises con más peliculas')
plt.xlabel('Cantidad')
plt.ylabel('Paises')
plt.show()
```
